#### Ejercicio 1: 

1.  Dados los siguientes datos: [15, 21, 25, 32, 35, 45, 50, 50, 55, 60]. Crea una funcion que calcule la media, mediana y moda.


```python
lista1=[15, 21, 25, 32, 35, 45, 50, 50, 55, 60]

def medidas_centrales(lista):
    datos=pd.DataFrame(lista)
    media=datos[0].mean()
    mediana=datos[0].median()
    moda=datos[0].mode()
    print(f"La media de los datos es {media}")
    print(f"La mediana de los datos es {mediana}")
    print(f"La moda de los datos es {moda}")

    
```

```python
medidas_centrales(lista1)
```

2.	Interpretación de la media: ¿Qué sucede si un valor extremo, como 100, se agrega al conjunto de datos anterior? ¿Cómo cambia la media? Explica por qué sucede esto.

```python
lista2=[15, 21, 25, 32, 35, 45, 50, 50, 55, 60,100]

medidas_centrales(lista2)
```

Podemos observar como en un primer momento la media se encontraba aproximadamnte 1,2 puntos por debajo de la mediana (mostrando que había bastantes valores pequeños los que desviaban la media a la baja).

En el segundo caso tras agregar el 100, esa diferencia entre media y mediana disminuyó hasta aproximadamente 0,6 puntos lo que tiene sentido con lo observado en el primer ejemplo. Si el valor extremo hubiera sido a la baja (un 1 por ejemplo) la diferencia entre media y mediana se hubiera acentuado aún más.
